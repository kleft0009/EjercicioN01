```python
i = 0

def func(x):
    global i
    i += 1
    return x**3 - 3*x**2 + x - 1

def secant_method(f, x0, x1, tol=1e-6, max_iter=100):
    x_prev, x_curr = x0, x1
    f_prev, f_curr = f(x_prev), f(x_curr)      
    iter_count = 0
    while abs(f_curr) > tol and iter_count < max_iter:
        x_next = x_curr - f_curr * (x_curr - x_prev) / (f_curr - f_prev)   
        f_next = f(x_next)                       
        x_prev, x_curr = x_curr, x_next
        f_prev, f_curr = f_curr, f_next          
        iter_count += 1
    return x_curr, iter_count

root, iters = secant_method(func, 2, 3)
print(f"Raíz: {root}, Iteraciones: {iters}, Llamadas a función: {i}")

```
